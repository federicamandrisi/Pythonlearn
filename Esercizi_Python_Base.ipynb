## 1. Stringhe

### Esempio 1: Concatenare stringhe
Puoi unire due stringhe usando l'operatore `+`.

```python
stringa1 = "Ciao"
stringa2 = "Mondo"
risultato = stringa1 + " " + stringa2
print(risultato)  # Output: Ciao Mondo
```

### Esercizio: Crea un messaggio di saluto
Unisci le stringhe "Benvenuto" e il tuo nome in un unico messaggio e stampalo.

```python
# Scrivi qui il tuo codice
str1 = "Benvenuta"
str2 = input("Scrivi il tuo nome:" )

# se la str2 è il tuo nome, allora stampalo, se è vuota dai errore
if str2 == "":
    print("Scrivi il tuo nome")
else:
    print(str1 + " " + str2)
```

## 2. Liste

### Esempio 2: Ordinare una lista
Puoi ordinare una lista con il metodo `.sort()`.

```python
numeri = [4, 2, 7, 1]
numeri.sort()
print(numeri)  # Output: [1, 2, 4, 7]
```

### Esercizio: Aggiungi e ordina
Aggiungi il numero `5` alla lista `[10, 2, 8]`, ordina la lista e stampala.

```python
# Scrivi qui il tuo codice
lista = [10, 2, 8]
lista.append(5)
lista.sort()
print(lista)
```

## 3. Float e Integer

### Esempio 3: Calcoli aritmetici
Puoi effettuare operazioni matematiche su numeri interi e float.

```python
numero1 = 7
numero2 = 3.5
somma = numero1 + numero2
print(somma)  # Output: 10.5
```

## 4. Dizionari

### Esempio 4: Accesso ai valori del dizionario
Puoi accedere ai valori di un dizionario usando le chiavi.

```python
studente = {"nome": "Luca", "età": 21}
print(studente['nome'])  # Output: Luca
print(studente['età'])  # Output: 21
```

### Esercizio: Aggiungi una chiave al dizionario
Dato un dizionario `auto = {'marca': 'Fiat', 'modello': 'Panda'}`, aggiungi la chiave `'anno'` con il valore `2022` e stampalo.

```python
# Scrivi qui il tuo codice
auto = {'marca': 'Fiat', 'modello': 'Panda'}
auto['anno'] = 2022
print(auto)
```
