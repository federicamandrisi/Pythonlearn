### Esercizio 1.1: Concatenare stringhe

### Esercizio 1.3: Manipolazione di stringhe
- Converti la stringa `"Python è divertente"` in maiuscolo.
- Sostituisci la parola `"divertente"` con `"utile"`.

```python
# Scrivi qui il tuo codice
stringa = "Python è divertente"
# Convertila in maiuscolo
# Sostituisci 'divertente' con 'utile'
stringa_sost = stringa.replace('divertente', 'utile')
stringa_maiuscolo = stringa_sost.upper()
print(stringa_maiuscolo)  # Output: PYTHON è UTILE
```

- Converti la prima parola 'Python' in maiuscolo.

```python
stringa = "Python è divertente"
parole = stringa.split()  # Dividi la stringa in parole
parole[0] = parole[0].upper()  # Trasforma la prima parola in maiuscolo
stringa_modificata = " ".join(parole)  # Ricostruisci la stringa
print(stringa_modificata)  # Output: "PYTHON è divertente"
```

- Converti l'ultima parola 'divertente' in maiuscolo.

#### APPROCCIO 1: Usare `.replace()`
Se la parola "divertente" compare una sola volta nella stringa, puoi sostituirla direttamente con la sua versione maiuscola.

```python
stringa = "Python è divertente"
stringa_modificata = stringa.replace("divertente", "DIVERTENTE")
print(stringa_modificata)  # Output: "Python è DIVERTENTE"
```

##### APPROCCIO 2: Manipolare la stringa tramite parole
Se preferisci un approccio basato su liste, puoi suddividere la stringa in parole e modificare solo quella specifica:

```python
stringa = "Python è divertente"
parole = stringa.split()  # Dividi la stringa in parole
for i, parola in enumerate(parole):
    if parola == "divertente":  # Trova la parola da modificare
        parole[i] = parola.upper()  # Trasforma in maiuscolo

stringa_modificata = " ".join(parole)  # Ricostruisci la stringa
print(stringa_modificata)  # Output: "Python è DIVERTENTE"
```

#### Differenza tra i due approcci:
- `.replace()` è più semplice ma sostituisce tutte le occorrenze della parola specificata.
- Divisione in parole è più flessibile se devi lavorare con più condizioni o modifiche parziali.

## 3. Float e Integer

### Esercizio 3.1: Operazioni matematiche
- Calcola il risultato di `7 * 3 + 5 / 2` e arrotondalo a 2 cifre decimali.

```python
# Scrivi qui il tuo codice
risultato = (7 * 3) + (5 / 2)
# Arrotonda il risultato a 2 cifre decimali
risultato_arrotondato = round(risultato, 2)
print(risultato_arrotondato)  # Output: 23.5
```

- Calcola il risultato di `(2 ** 3) - 4 / 2` e arrotondalo a 2 cifre decimali.

```python
risultato = (2 ** 3) - 4 / 3 #Output : 6.6 periodically
 # Arrotonda il risultato a 2 cifre decimali
risultato_arrotondato = round(risultato, 2)
print(risultato_arrotondato) # Output: 6.0
```

### Esercizio 3.2: Conversioni
Converta il numero `3.9` in un intero e poi il numero `4` in un float.

```python
# Scrivi qui il tuo codice
numero_float = 3.9
numero_intero = 4
# Convertili nei rispettivi tipi
num_intero_converted = int(numero_float)
num_float_converted = float(numero_intero)
print(num_intero_converted)  # Output: 3
print(num_float_converted)  # Output: 4.0
```

## 4. Dizionari

### Esercizio 4.1: Iterazioni sui dizionari
Dato il dizionario `auto = {'marca': 'Fiat', 'modello': 'Panda', 'anno': 2022}`:
- Stampa tutte le chiavi.
- Stampa tutti i valori.

```python
# Scrivi qui il tuo codice
auto = {'marca': 'Fiat', 'modello': 'Panda', 'anno': 2022}
# Stampa chiavi e valori
for chiave in auto.keys():
    print(chiave)
for valore in auto.values():
    print(valore)
```

### Esercizio 4.2: Aggiornare un dizionario
Aggiungi al dizionario `studente = {'nome': 'Luca', 'età': 21}` la chiave `'corso'` con valore `'Informatica'`.

```python
# Scrivi qui il tuo codice
studente = {'nome': 'Luca', 'età': 21}
# Aggiungi la chiave 'corso'
studente['corso'] = 'Informatica'
print(studente)  # Output: {'nome': 'Luca', 'età': 21, 'corso': 'Informatica'}
```
